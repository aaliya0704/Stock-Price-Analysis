```python
import pandas as pd 
import matplotlib.pyplot as plt 
import yfinance as yf
```

```python
df = yf.download("AAPL")
```

```python
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(12,6))

plt.plot(df.index, df["Close"])

plt.title("Apple Stock Closing Price")

plt.xlabel("Date")

plt.ylabel("Price (USD)")

plt.show()
```

```python
df["Close"].rolling(window=20).mean()
```

```python
df["MA20"] = df["Close"].rolling(window=20).mean()
```

```python
df["MA20"]
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df["Close"], label="Close")
plt.plot(df.index, df["MA20"], label="20-day moving average")
plt.title("Apple stock prices with moving average")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
df["Daily change"] = df["Close"] - df["Open"]
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df["Daily change"])
plt.title("Daily Price change")
plt.xlabel("Date")
plt.ylabel("Price Change")
plt.show()
```

```python
ticker = yf.Ticker("AAPL")

df = ticker.history(period="1y")
```

```python
df.head()
```

```python
df.info()
```

```python
df["Close"].head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df["Close"])
plt.title("Apple stock closing price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python
df["MA20"] = df["Close"].rolling(window=20).mean()
```

```python
df[["Close","MA20"]].head(25)
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df["Close"], label="Close")
plt.plot(df.index, df["MA20"], label="20-day moving average")
plt.title("Apple stock price with moving average")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
df["Daily_Return"]= df["Close"].pct_change()
```

```python
df["Daily_Return"].head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df["Daily_Return"])
plt.title("Apple Daily Returns")
plt.xlabel("Date")
plt.ylabel("Return")
plt.show()
```
